```python
#data load

import pandas as pd

df=pd.read_excel(r"E:\Down\analyst\sales_raw_500.xlsx")
```

```python
df
```

```python
#duplicate removes

df = df.drop_duplicates()
```

```python
df
```

```python
#fix date column

df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

```python
df = df.dropna(subset=['Order_Date'])
```

```python
df
```

```python
df['Profit'] = df['Sales']-df['Cost']
df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
df
```

```python
df.to_csv(r"E:\Down\analyst\sales_cleaned.csv", index=False)
```
